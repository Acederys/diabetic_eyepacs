```python
# Преобразуем в список строк
records = []
for model, folds in results.items():
    for fold, metrics in folds.items():
        row = {'model': model, 'fold': fold}
        row.update(metrics)
        records.append(row)

df = pd.DataFrame(records)
```

```python
df
```

```python
df['block_id'] = df.groupby('fold').ngroup()
```

```python
import matplotlib.pyplot as plt
import scikit_posthocs as sp
import scipy.stats as ss

metrics = ['f1_macro', 'f1_weighted', 'kappa', 'auc_roc', 'recall_1', 'recall_2']

fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(20, 24), dpi=150)
fig.suptitle('Friedman + Conover post-hoc + CD Diagram per Metric', fontsize=16)

for i, metric in enumerate(metrics):
    # Friedman test
    grouped = [df[df['model'] == model][metric].values for model in df['model'].unique()]
    stat, p = ss.friedmanchisquare(*grouped)
    print(f"{metric}: Friedman χ² = {stat:.4f}, p = {p:.4g}")

    # Пост-хок Conover-Friedman
    test_results = sp.posthoc_conover_friedman(
        df[['model', 'fold', metric, 'block_id']].rename(columns={metric: 'score'}),
        melted=True,
        block_col='fold',
        block_id_col='block_id',
        group_col='model',
        y_col='score',
    )

    # График значимости
    sp.sign_plot(test_results, ax=axes[i, 0])
    axes[i, 0].set_title(f'Significance plot: {metric}')

    # Диаграмма критической разницы
    avg_rank = df.groupby('model')[metric].mean().rank()
    sp.critical_difference_diagram(avg_rank, test_results, ax=axes[i, 1])
    axes[i, 1].set_title(f'Critical Difference: {metric}')

plt.tight_layout(rect=[0, 0.03, 0.95, 0.97])
plt.show()
```
