```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Данные
data = {
    'Metric': ['Precision',
'Recall',
'F1-Score',
'accuracy',
'Kappa',
'AUC-ROC'],
    'fold 1': [0.6039,0.6414,0.6037, 0.7848, 0.4145,0.8349],
    'fold 2': [0.6144,0.6530,0.6109, 0.7919, 0.4313,0.8359],
    'fold 3': [0.6115,0.6425,0.6041,0.7884, 0.4157,0.8339]
}

# Преобразуем в DataFrame
df = pd.DataFrame(data)

# Переводим в длинный формат для seaborn
df_melted = df.melt(id_vars='Metric', value_vars=['fold 1', 'fold 2', 'fold 3'], var_name='Model', value_name='Score')

# Визуализация
plt.figure(figsize=(8, 6))
sns.barplot(x='Metric', y='Score', hue='Model', data=df_melted, palette='Blues')

plt.title('Model Performance Comparison', fontsize=16)
plt.ylabel('Score')
plt.xlabel('Metrics')
plt.ylim(0, 1)  # Ограничиваем шкалу для наглядности
plt.tight_layout()
plt.show()

# агментация 2 класса
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# базовая модель 
models = {
    'fold 1':      np.array([[4833,  329,    0],
 [ 977,  745,    0],
 [  20 , 122 ,   0],
]),
    'fold 2': np.array([[4836,  326,    0],
 [ 947,  775,    0],
 [  20,  122 ,   0],
]),
    'fold 3':           np.array([[4828,  334,    0],
 [ 951,  771,    0],
 [  25,  114 ,   3],
])
}

fig, axs = plt.subplots(1, 3, figsize=(15, 4))

for ax, (name, matrix) in zip(axs, models.items()):
    sns.heatmap(matrix, annot=True, fmt='d', cmap='Greens', xticklabels=['NORMA', 'NPDR', 'PDR'], yticklabels=['NORMA', 'NPDR', 'PDR'], ax=ax)
    ax.set_title(name)
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

plt.tight_layout()
plt.show()
```
